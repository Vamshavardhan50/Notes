```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
df = sns.load_dataset("iris")
```

```python
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df["species"].value_counts()
```

```python
X=df.drop("species", axis=1)
y=df["species"]
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.33,random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
model_knn = KNeighborsClassifier(n_neighbors=13)
```

```python
model_knn.fit(X_train,y_train)
```

```python
y_pred  =model_knn.predict(X_test)
```

```python
y_pred 
```

```python
model_knn.score(X_test,y_test)
```

```python
from sklearn.svm import SVC

model_SVM = SVC(gamma='auto')  
```

```python
model_SVM.fit(X_train,y_train)
model_SVM.score(X_test,y_test)
```

```python
#now lets we do GridSearchCv
from sklearn.model_selection import GridSearchCV
```

```python
classifier = GridSearchCV((model_SVM),{'C': [1, 1, 10, 100],'kernel': ['linear', 'rbf', 'poly']},cv=5,return_train_score=False)
```

```python
classifier.fit(X,y)
```

```python
classifier.cv_results_
```

```python
results = pd.DataFrame(classifier.cv_results_)
```

```python
results   
```

```python
results[['param_C', 'param_kernel', 'mean_test_score']]
```

```python
from sklearn.model_selection import RandomizedSearchCV
classifier_2 =RandomizedSearchCV((model_SVM),{
  'C':[1,10,20,30],
  'kernel':['linear', 'rbf', 'poly']
},cv=5,return_train_score=False,n_iter=10)
```

```python
classifier_2.fit(X,y)
```

```python
classifier_2.cv_results_
```
