```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns 
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import make_blobs
```

```python
X,y = make_blobs(n_samples=500, n_features=5,centers=3, cluster_std=1.5, random_state=42)
```

```python
X  
```

```python
y
```

```python
df = pd.DataFrame(X, columns=[f'Feature {i+1}' for i in range(X.shape[1])])
```

```python
df
```

```python
scaler  = StandardScaler()
X_scaled = scaler.fit_transform(df)
```

```python
X_scaled
```

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
```

```python
X_pca
```

```python
df = pd.DataFrame(X_pca, columns=['PC1', 'PC2'])
df['labels'] = y
```

```python
df  
```

```python
sns.scatterplot(data=df, x='PC1', y='PC2', hue='labels', palette='viridis')
```
