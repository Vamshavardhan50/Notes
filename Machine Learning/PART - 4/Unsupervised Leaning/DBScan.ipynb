```python
import pandas as pd 
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans,DBSCAN
from sklearn.preprocessing import StandardScaler
import seaborn as sns
from sklearn.datasets import make_moons
```

```python
X, y = make_moons(n_samples=300,noise=0.06 , random_state=42)
```

```python
X  
```

```python
df = pd.DataFrame(X, columns=['Feature1', 'Feature2'])
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df)
```

```python
kmeans = KMeans(n_clusters=2, random_state=42)
kmeans_label = kmeans.fit_predict(X_scaled)
```

```python
df['kmeans_clusters'] = kmeans_label
```

```python
df  
```

```python
sns.scatterplot(data=df, x='Feature1', y='Feature2', hue='kmeans_clusters', palette='Set1', s=100) 
```

```python
dbscan = DBSCAN(eps=0.2, min_samples=5)
dbscan_label = dbscan.fit_predict(X_scaled)
```

```python
df['dbscan_clusters'] = dbscan_label
```

```python
df  
```

```python
sns.scatterplot(data=df, x='Feature1', y='Feature2', hue='dbscan_clusters', palette='viridis')
```
