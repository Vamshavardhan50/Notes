We use **distribution plots** to visualize the distribution of quantitative data.

Let's discuss some plots that allow us to visualize the distribution of a data set.  
These plots are:

- `distplot` / `histplot`
- `jointplot`
- `pairplot`
- `rugplot`


```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=sns.load_dataset("tips")
df.head()
```

```python
plt.subplot(1,2,1)
sns.histplot(df["total_bill"],bins=30,kde=True)
plt.subplot(1,2,2)
sns.histplot(df["tip"],bins=30,kde=True)
```

```python
#join plot
sns.jointplot(x="total_bill", y="tip", data=df, kind="reg")
```

```python
#pair plot
sns.pairplot(df,hue="day")
```

```python
#rugplot
sns.rugplot(df["total_bill"])
```
