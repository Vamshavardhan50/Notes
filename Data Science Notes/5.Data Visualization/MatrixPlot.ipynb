```python
import seaborn as sns
import matplotlib.pyplot as plt
```

**Heat Map**

```python
flights = sns.load_dataset("flights")
flights
```

```python
tips =sns.load_dataset("tips")
tips 
```

```python
tipscorr = tips[["total_bill", "tip", "size"]].corr()
sns.heatmap(tipscorr,annot=True, cmap="coolwarm")
```

**Pivot Table Heat map**

```python

pivot_flights = flights.pivot_table(values="passengers",index="month", columns="year")
pivot_flights
```

```python
sns.heatmap(pivot_flights)
```
